```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
precisions = ['amp_','basic_','half_']
halfs = ['','50_']
batch_sizes = ['64','128']
```

```python
%cd '/content/drive/MyDrive/HPML/project/ddp4'
fig, ax = plt.subplots(2,2,figsize=(15,10))
for precision in precisions:
  for half in halfs:
    batch_size='64'
    file_name = precision+half+'005_'+batch_size+'_ddp4.txt'
    print(file_name)
    data = np.loadtxt(file_name)   # 600 x 10

    ax[0][0].plot(data[:,0],(data[:,6]+data[:,7]+data[:,8]+data[:,9])/400,label=file_name)
    ax[0][0].set_xlabel('epoch')
    ax[0][0].set_ylabel('loss for 4 devices')
    ax[0][1].plot(data[:,0],(data[:,2]+data[:,3]+data[:,4]+data[:,5])/4)
    ax[0][1].set_xlabel('epoch')
    ax[0][1].set_ylabel('acc for 4 devices')

    ax[1][0].plot(data[:,1],(data[:,6]+data[:,7]+data[:,8]+data[:,9])/400)
    ax[1][0].set_xlabel('total time')
    ax[1][0].set_ylabel('loss for 4 devices')
    ax[1][1].plot(data[:,1],(data[:,2]+data[:,3]+data[:,4]+data[:,5])/4)
    ax[1][1].set_xlabel('total time')
    ax[1][1].set_ylabel('acc for 4 devices')
    fig.legend()
    break

plt.savefig('ddp64.jpg')
plt.show()
```
